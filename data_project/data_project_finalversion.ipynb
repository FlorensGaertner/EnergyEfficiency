##### Surface Area (Oberfläche):
###### • Die Gesamtfläche aller äußeren Oberflächen eines Gebäudes, einschließlich Wände, Dach und Boden.
##### Wall Area (Wandfläche):
###### • Die Gesamtfläche der Außenwände eines Gebäudes.
##### Roof Area (Dachfläche):
###### • Die Gesamtfläche des Daches eines Gebäudes.
##### Overall Height (Gesamthöhe):
###### • Die maximale vertikale Höhe eines Gebäudes, gemessen vom Boden bis zum höchsten Punkt des Daches.
##### Orientation (Ausrichtung):
###### • Die Himmelsrichtung, in die die Hauptfassade eines Gebäudes zeigt (z.B. Norden, Süden, Osten, Westen).
##### Glazing Area (Verglasungsfläche):
###### • Die Gesamtfläche aller Fenster und Glasflächen in einem Gebäude.
##### Glazing Area Distribution (Verteilung der Verglasungsfläche):
###### • Die Art und Weise, wie die Verglasungsfläche über die verschiedenen Seiten eines Gebäudes verteilt ist (z.B. gleichmäßig oder ungleichmäßig auf verschiedene Fassaden verteilt).
##### Heating Load (Heizlast):
###### • Die Menge an Wärmeenergie, die benötigt wird, um ein Gebäude in den kälteren Monaten auf einer komfortablen Temperatur zu halten.
##### Cooling Load (Kühllast):
###### • Die Menge an Kälteenergie, die benötigt wird, um ein Gebäude in den wärmeren Monaten auf einer komfortablen Temperatur zu halten.

Outlier Searcher

```python
import pandas as pd

# Load data
df = pd.read_csv("energy_efficiency_data.csv")
df.columns = [col.replace("AAPL.", "") for col in df.columns]

# Iterate over each column to check for outliers
for col in df.columns:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define outlier criteria
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Find outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    # Print outliers for each column
    if not outliers.empty:
        print(f"Outliers in {col}:")
        print(outliers)
    else:
        print(f"No outliers found in {col}")

```

```python
import plotly.graph_objects as go
import pandas as pd

# Load data
df = pd.read_csv("energy_efficiency_data.csv")
df.columns = [col.replace("AAPL.", "") for col in df.columns]

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['Heating_Load'].quantile(0.25)
Q3 = df['Heating_Load'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier criteria
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(df['Heating_Load'] < lower_bound) | (df['Heating_Load'] > upper_bound)]
non_outliers = df[(df['Heating_Load'] >= lower_bound) & (df['Heating_Load'] <= upper_bound)]

# Create figure
fig = go.Figure()

# Add non-outlier data
fig.add_trace(
    go.Scatter(
        x=non_outliers["Surface_Area"], 
        y=non_outliers["Heating_Load"], 
        mode='markers', 
        name='Non-Outliers'
    )
)

# Add outlier data
fig.add_trace(
    go.Scatter(
        x=outliers["Surface_Area"], 
        y=outliers["Heating_Load"], 
        mode='markers', 
        marker=dict(color='red', size=10, symbol='x'), 
        name='Outliers'
    )
)

# Set title and layout
fig.update_layout(
    title="Heating Load dependent on Surface Area (with Outliers)",
    xaxis=dict(
        rangeslider=dict(
            visible=True
        )
    ),
    xaxis_title="Surface Area",
    yaxis_title="Heating Load"
)

fig.show()
```

import plotly.graph_objects as go
import pandas as pd

# Load data
df = pd.read_csv("energy_efficiency_data.csv")
df.columns = [col.replace("AAPL.", "") for col in df.columns]

# Create a function to generate vertices for a house given its dimensions
def generate_house_vertices(wall_area, roof_area, height):
    length = wall_area ** 0.5  # Simplifying assumption: wall area forms a square for simplicity
    width = roof_area ** 0.5  # Simplifying assumption: roof area forms a square for simplicity
    return [
        (0, 0, 0), (length, 0, 0), (length, width, 0), (0, width, 0),  # Base vertices
        (0, 0, height), (length, 0, height), (length, width, height), (0, width, height)  # Top vertices
    ]

# Create figure
fig = go.Figure()

for i, row in df.iterrows():
    wall_area = row['Wall_Area']
    roof_area = row['Roof_Area']
    height = row['Surface_Area'] / wall_area  # Simplified height calculation
    
    vertices = generate_house_vertices(wall_area, roof_area, height)
    
    # Define edges by connecting vertices
    edges = [
        (0, 1), (1, 2), (2, 3), (3, 0),  # Base edges
        (4, 5), (5, 6), (6, 7), (7, 4),  # Top edges
        (0, 4), (1, 5), (2, 6), (3, 7)  # Vertical edges
    ]
    
    for edge in edges:
        x_coords = [vertices[edge[0]][0], vertices[edge[1]][0]]
        y_coords = [vertices[edge[0]][1], vertices[edge[1]][1]]
        z_coords = [vertices[edge[0]][2], vertices[edge[1]][2]]
        
        fig.add_trace(go.Scatter3d(
            x=x_coords, y=y_coords, z=z_coords,
            mode='lines',
            line=dict(color='blue'),
            showlegend=False
        ))

# Set titles and layout
fig.update_layout(
    title="3D Representation of Houses Based on Wall Area, Roof Area, and Height",
    scene=dict(
        xaxis_title='Length',
        yaxis_title='Width',
        zaxis_title='Height'
    ),
    margin=dict(l=0, r=0, b=0, t=40)
)
