##### Surface Area (Oberfläche):
###### • Die Gesamtfläche aller äußeren Oberflächen eines Gebäudes, einschließlich Wände, Dach und Boden.
##### Wall Area (Wandfläche):
###### • Die Gesamtfläche der Außenwände eines Gebäudes.
##### Roof Area (Dachfläche):
###### • Die Gesamtfläche des Daches eines Gebäudes.
##### Overall Height (Gesamthöhe):
###### • Die maximale vertikale Höhe eines Gebäudes, gemessen vom Boden bis zum höchsten Punkt des Daches.
##### Orientation (Ausrichtung):
###### • Die Himmelsrichtung, in die die Hauptfassade eines Gebäudes zeigt (z.B. Norden, Süden, Osten, Westen).
##### Glazing Area (Verglasungsfläche):
###### • Die Gesamtfläche aller Fenster und Glasflächen in einem Gebäude.
##### Glazing Area Distribution (Verteilung der Verglasungsfläche):
###### • Die Art und Weise, wie die Verglasungsfläche über die verschiedenen Seiten eines Gebäudes verteilt ist (z.B. gleichmäßig oder ungleichmäßig auf verschiedene Fassaden verteilt).
##### Heating Load (Heizlast):
###### • Die Menge an Wärmeenergie, die benötigt wird, um ein Gebäude in den kälteren Monaten auf einer komfortablen Temperatur zu halten.
##### Cooling Load (Kühllast):
###### • Die Menge an Kälteenergie, die benötigt wird, um ein Gebäude in den wärmeren Monaten auf einer komfortablen Temperatur zu halten.

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Assuming df is your DataFrame
# df = pd.read_csv('your_data.csv')
df = pd.read_csv("energy_efficiency_data.csv")  #Defining the variable 'df' for our data

# Create subplots with 1 row and 2 columns
fig = make_subplots(rows=1, cols=2, subplot_titles=('Heating Load', 'Cooling Load'))

# First scatter plot - Heating Load vs Surface Area
fig.add_trace(
    go.Scatter(
        x=df['Surface_Area'], 
        y=df['Heating_Load'], 
        mode='markers', 
        marker=dict(size=7, color=df['Heating_Load'], colorscale='Plasma', opacity=0.7, colorbar=dict(x=0.45, title='Heating Load'))
    ),
    row=1, col=1
)

# Second scatter plot - Cooling Load vs Surface Area
fig.add_trace(
    go.Scatter(
        x=df['Surface_Area'], 
        y=df['Cooling_Load'], 
        mode='markers', 
        marker=dict(size=8, color=df['Cooling_Load'], colorscale='Plasma', opacity=0.7, colorbar=dict(title='Cooling Load'))
    ),
    row=1, col=2
)

# Update xaxis and yaxis labels for both plots
fig.update_xaxes(title_text='Surface Area', row=1, col=1)
fig.update_yaxes(title_text='Heating Load', row=1, col=1)

fig.update_xaxes(title_text='Surface Area', row=1, col=2)
fig.update_yaxes(title_text='Cooling Load', row=1, col=2)

# Update layout
fig.update_layout(
    title_text='Heating Load and Cooling Load vs Surface Area',
    showlegend=False,
    plot_bgcolor='rgba(0,0,0,0)'  # Setting plot background to transparent
)

# Adjust background colors for the scatter plots
fig.update_scenes(
    bgcolor='rgba(255, 0, 0, 0.1)',  # Reddish background for Heating Load plot
    row=1, col=1
)
fig.update_scenes(
    bgcolor='rgba(0, 0, 255, 0.1)',  # Bluish background for Cooling Load plot
    row=1, col=2
)

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# Load data
df = pd.read_csv("energy_efficiency_data.csv")
df.columns = [col.replace("AAPL.", "") for col in df.columns]

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['Heating_Load'].quantile(0.25)
Q3 = df['Heating_Load'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier criteria
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(df['Heating_Load'] < lower_bound) | (df['Heating_Load'] > upper_bound)]
non_outliers = df[(df['Heating_Load'] >= lower_bound) & (df['Heating_Load'] <= upper_bound)]

# Create figure
fig = go.Figure()

# Add non-outlier data
fig.add_trace(
    go.Scatter(
        x=non_outliers["Surface_Area"], 
        y=non_outliers["Heating_Load"], 
        mode='markers', 
        name='Non-Outliers'
    )
)

# Add outlier data
fig.add_trace(
    go.Scatter(
        x=outliers["Surface_Area"], 
        y=outliers["Heating_Load"], 
        mode='markers', 
        marker=dict(color='red', size=10, symbol='x'), 
        name='Outliers'
    )
)

# Set title and layout
fig.update_layout(
    title="Heating Load dependent on Surface Area (with Outliers)",
    xaxis=dict(
        rangeslider=dict(
            visible=True
        )
    ),
    xaxis_title="Surface Area",
    yaxis_title="Heating Load"
)

fig.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# Assuming df is your DataFrame and it has the columns 'Glazing_Area', 'Heating_Load', and 'Cooling_Load'
df = pd.read_csv("energy_efficiency_data.csv")

# Create horizontal boxplot for Heating Load
fig_heating_glazing = px.box(df, y="Glazing_Area", x="Heating_Load",
                             labels={"Heating_Load": "Heating Load", "Glazing_Area": "Glazing Area"},
                             orientation='h')

# Create horizontal boxplot for Cooling Load
fig_cooling_glazing = px.box(df, y="Glazing_Area", x="Cooling_Load",
                             labels={"Cooling_Load": "Cooling Load", "Glazing_Area": "Glazing Area"},
                             orientation='h')

# Create subplots
fig2 = make_subplots(rows=1, cols=2, subplot_titles=("Impact of Glazing Area on Heating Load",
                                                     "Impact of Glazing Area on Cooling Load"),
                     horizontal_spacing=0.1)

# Add Heating Load plot
for trace in fig_heating_glazing['data']:
    fig2.add_trace(trace, row=1, col=1)

# Add Cooling Load plot
for trace in fig_cooling_glazing['data']:
    fig2.add_trace(trace, row=1, col=2)

# Define custom colors for the box plots
colors = ['#636EFA', '#EF553B', '#00CC96', '#AB63FA']

# Apply custom colors to the box plots
for i, trace in enumerate(fig2.data):
    if i < len(fig_heating_glazing['data']):
        trace.marker.color = colors[i % len(colors)]  # Apply color for Heating Load
    else:
        trace.marker.color = colors[(i - len(fig_heating_glazing['data'])) % len(colors)]  # Apply color for Cooling Load

# Update layout
fig2.update_layout(title_text="Impact of Glazing Area on Heating and Cooling Load",
                   showlegend=False)

# Show plot
fig2.show()
```

import plotly.graph_objects as go
import pandas as pd

# Load data
df = pd.read_csv("energy_efficiency_data.csv")
df.columns = [col.replace("AAPL.", "") for col in df.columns]

# Create a function to generate vertices for a house given its dimensions
def generate_house_vertices(wall_area, roof_area, height):
    length = wall_area ** 0.5  # Simplifying assumption: wall area forms a square for simplicity
    width = roof_area ** 0.5  # Simplifying assumption: roof area forms a square for simplicity
    return [
        (0, 0, 0), (length, 0, 0), (length, width, 0), (0, width, 0),  # Base vertices
        (0, 0, height), (length, 0, height), (length, width, height), (0, width, height)  # Top vertices
    ]

# Create figure
fig = go.Figure()

for i, row in df.iterrows():
    wall_area = row['Wall_Area']
    roof_area = row['Roof_Area']
    height = row['Surface_Area'] / wall_area  # Simplified height calculation
    
    vertices = generate_house_vertices(wall_area, roof_area, height)
    
    # Define edges by connecting vertices
    edges = [
        (0, 1), (1, 2), (2, 3), (3, 0),  # Base edges
        (4, 5), (5, 6), (6, 7), (7, 4),  # Top edges
        (0, 4), (1, 5), (2, 6), (3, 7)  # Vertical edges
    ]
    
    for edge in edges:
        x_coords = [vertices[edge[0]][0], vertices[edge[1]][0]]
        y_coords = [vertices[edge[0]][1], vertices[edge[1]][1]]
        z_coords = [vertices[edge[0]][2], vertices[edge[1]][2]]
        
        fig.add_trace(go.Scatter3d(
            x=x_coords, y=y_coords, z=z_coords,
            mode='lines',
            line=dict(color='blue'),
            showlegend=False
        ))

# Set titles and layout
fig.update_layout(
    title="3D Representation of Houses Based on Wall Area, Roof Area, and Height",
    scene=dict(
        xaxis_title='Length',
        yaxis_title='Width',
        zaxis_title='Height'
    ),
    margin=dict(l=0, r=0, b=0, t=40)
)

# Show plot
fig.show()
