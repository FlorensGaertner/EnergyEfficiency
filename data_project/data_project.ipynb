##### Surface Area (Oberfläche):
###### • Die Gesamtfläche aller äußeren Oberflächen eines Gebäudes, einschließlich Wände, Dach und Boden.
##### Wall Area (Wandfläche):
###### • Die Gesamtfläche der Außenwände eines Gebäudes.
##### Roof Area (Dachfläche):
###### • Die Gesamtfläche des Daches eines Gebäudes.
##### Overall Height (Gesamthöhe):
###### • Die maximale vertikale Höhe eines Gebäudes, gemessen vom Boden bis zum höchsten Punkt des Daches.
##### Orientation (Ausrichtung):
###### • Die Himmelsrichtung, in die die Hauptfassade eines Gebäudes zeigt (z.B. Norden, Süden, Osten, Westen).
##### Glazing Area (Verglasungsfläche):
###### • Die Gesamtfläche aller Fenster und Glasflächen in einem Gebäude.
##### Glazing Area Distribution (Verteilung der Verglasungsfläche):
###### • Die Art und Weise, wie die Verglasungsfläche über die verschiedenen Seiten eines Gebäudes verteilt ist (z.B. gleichmäßig oder ungleichmäßig auf verschiedene Fassaden verteilt).
##### Heating Load (Heizlast):
###### • Die Menge an Wärmeenergie, die benötigt wird, um ein Gebäude in den kälteren Monaten auf einer komfortablen Temperatur zu halten.
##### Cooling Load (Kühllast):
###### • Die Menge an Kälteenergie, die benötigt wird, um ein Gebäude in den wärmeren Monaten auf einer komfortablen Temperatur zu halten.

```python
import plotly.graph_objects as go
import pandas as pd

# Load data
df = pd.read_csv("energy_efficiency_data.csv")
df.columns = [col.replace("AAPL.", "") for col in df.columns]

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['Heating_Load'].quantile(0.25)
Q3 = df['Heating_Load'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier criteria
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(df['Heating_Load'] < lower_bound) | (df['Heating_Load'] > upper_bound)]
non_outliers = df[(df['Heating_Load'] >= lower_bound) & (df['Heating_Load'] <= upper_bound)]

# Create figure
fig = go.Figure()

# Add non-outlier data
fig.add_trace(
    go.Scatter(
        x=non_outliers["Surface_Area"], 
        y=non_outliers["Heating_Load"], 
        mode='markers', 
        name='Non-Outliers'
    )
)

# Add outlier data
fig.add_trace(
    go.Scatter(
        x=outliers["Surface_Area"], 
        y=outliers["Heating_Load"], 
        mode='markers', 
        marker=dict(color='red', size=10, symbol='x'), 
        name='Outliers'
    )
)

# Set title and layout
fig.update_layout(
    title="Heating Load dependent on Surface Area (with Outliers)",
    xaxis=dict(
        rangeslider=dict(
            visible=True
        )
    ),
    xaxis_title="Surface Area",
    yaxis_title="Heating Load"
)

fig.show()
```

# Load data
df = pd.read_csv("energy_efficiency_data.csv")
df.columns = [col.replace("AAPL.", "") for col in df.columns]

# Create a function to generate vertices for a house given its dimensions
def generate_house_vertices(wall_area, roof_area, height):
    length = wall_area ** 0.5  # Simplifying assumption: wall area forms a square for simplicity
    width = roof_area ** 0.5  # Simplifying assumption: roof area forms a square for simplicity
    return [
        (0, 0, 0), (length, 0, 0), (length, width, 0), (0, width, 0),  # Base vertices
        (0, 0, height), (length, 0, height), (length, width, height), (0, width, height)  # Top vertices
    ]

# Create figure
fig = go.Figure()

for i, row in df.iterrows():
    wall_area = row['Wall_Area']
    roof_area = row['Roof_Area']
    height = row['Surface_Area'] / wall_area  # Simplified height calculation
    
    vertices = generate_house_vertices(wall_area, roof_area, height)
    
    # Define edges by connecting vertices
    edges = [
        (0, 1), (1, 2), (2, 3), (3, 0),  # Base edges
        (4, 5), (5, 6), (6, 7), (7, 4),  # Top edges
        (0, 4), (1, 5), (2, 6), (3, 7)  # Vertical edges
    ]
    
    for edge in edges:
        x_coords = [vertices[edge[0]][0], vertices[edge[1]][0]]
        y_coords = [vertices[edge[0]][1], vertices[edge[1]][1]]
        z_coords = [vertices[edge[0]][2], vertices[edge[1]][2]]
        
        fig.add_trace(go.Scatter3d(
            x=x_coords, y=y_coords, z=z_coords,
            mode='lines',
            line=dict(color='blue'),
            showlegend=False
        ))

# Set titles and layout

```python
import pandas as pd
from dash import Dash, dcc, html
import plotly.express as px
from dash.dependencies import Input, Output

# Specify full path for the data file
data_file = r'C:\Users\flore\OneDrive\Desktop\data_project\energy_efficiency_data.csv'

# Load your data
df = pd.read_csv(data_file)

# Calculate Total_Load and drop Heating_Load, Cooling_Load columns if they exist
if 'Heating_Load' in df.columns and 'Cooling_Load' in df.columns:
    df['Total_Load'] = df['Heating_Load'] + df['Cooling_Load']
    df_for_scatter = df.copy()  # Make a copy for scatter plot and table
    df_for_scatter.drop(['Heating_Load', 'Cooling_Load', 'Wall_Area', 'Roof_Area', 'Overall_Height'], axis=1, inplace=True)
else:
    df_for_scatter = df.drop(['Wall_Area', 'Roof_Area', 'Overall_Height'], axis=1)

# Calculate mean Total_Load per Relative_Compactness and Glazing_Area
mean_total_load_per_rc_ga = df_for_scatter.groupby(['Relative_Compactness', 'Glazing_Area'])['Total_Load'].mean().reset_index()
mean_total_load_per_rc_ga['Total_Load'] = mean_total_load_per_rc_ga['Total_Load'].round(1)  # Round to 1 decimal place
mean_total_load_per_rc_ga['Total_Load'] = mean_total_load_per_rc_ga['Total_Load'].apply(lambda x: f'{x:.1f}' if x.is_integer() else f'{x}')

# Initialize the Dash app
app = Dash(__name__)

# Define app layout with inline styles for larger width
app.layout = html.Div(
    style={'width': '90%', 'max-width': '1200px', 'margin': 'auto'},  # Adjust width and max-width for larger dashboard
    children=[
        html.H1("Energy Efficiency Dashboard"),  # Header displayed at the top of the dashboard

        # Heatmap for correlation matrix
        html.Div([
            html.H2("Correlation Matrix of Energy Efficiency Factors"),
            dcc.Graph(
                id='correlation-heatmap',
                figure={}  # Placeholder for now
            )
        ]),

        html.Hr(),

        # Button group to select Relative_Compactness with larger buttons
        dcc.RadioItems(
            id='compactness-radio',
            options=[{'label': val, 'value': val} for val in df['Relative_Compactness'].unique()],
            value=df['Relative_Compactness'].unique()[0],  # Set default value to the first option
            labelStyle={'display': 'inline-block', 'margin-right': '10px', 'font-size': '18px'}  # Adjust font-size for larger buttons
        ),

        # Graph for Total_Load vs Glazing_Area
        dcc.Graph(id='total-load-graph'),

        # Table to display mean Total_Load per Glazing_Area
        html.Div(id='mean-load-table'),
    ]
)

# Define callback to update graph, table, and heatmap based on radio button selection
@app.callback(
    [Output('total-load-graph', 'figure'),
     Output('mean-load-table', 'children'),
     Output('correlation-heatmap', 'figure')],
    [Input('compactness-radio', 'value')]
)
def update_data(selected_compactness):
    # Filter data for the selected Relative_Compactness for scatter plot and table
    filtered_df = mean_total_load_per_rc_ga[mean_total_load_per_rc_ga['Relative_Compactness'] == selected_compactness]
    
    # Scatter plot figure with colorscale for y-axis
    fig = px.scatter(filtered_df, x='Glazing_Area', y='Total_Load', color='Total_Load', color_continuous_scale='Viridis',
                     hover_data=['Relative_Compactness', 'Glazing_Area', 'Total_Load'],
                     title=f'Mean Total Load per Glazing Area ({selected_compactness})')
    
    # Set fixed range for y-axis and synchronize with x-axis
    fig.update_yaxes(fixedrange=True, scaleanchor='x')
    
    # Reverse y-axis to display higher values at the top
    fig.update_layout(yaxis=dict(autorange='reversed'))
    
    # Table with mean Total_Load per Glazing_Area
    table = html.Table([
        html.Thead(
            html.Tr([html.Th(col) for col in filtered_df.columns])
        ),
        html.Tbody([
            html.Tr([
                html.Td(str(filtered_df.iloc[i][col])) for col in filtered_df.columns
            ]) for i in range(len(filtered_df))
        ])
    ])
    
    # Correlation matrix for heatmap (including all columns except Total_Load)
    correlation_matrix = df.corr()
    heatmap_fig = px.imshow(correlation_matrix,
                            labels=dict(color='Correlation'),
                            color_continuous_scale='Viridis',
                            zmin=-1, zmax=1).update_layout(title="Correlation Heatmap")
    
    return fig, table, heatmap_fig

# Run the app in Jupyter Notebook cell with 'inline' mode
app.run_server(mode='inline', debug=True)
```
